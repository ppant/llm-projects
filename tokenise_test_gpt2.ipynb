```python
from transformers import GPT2Tokenizer
def tokenize_for_gpt2(text, tokenizer_name="gpt2"):
    tokenizer = GPT2Tokenizer.from_pretrained(tokenizer_name)
    return tokenizer.encode(text, return_tensors="pt")  # Return PyTorch tensors
tokenize_for_gpt2("This is new text which needs to be tokenise")
```
